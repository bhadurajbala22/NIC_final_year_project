```python
#from skimage.measure import 
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2
```

```python
def mse(imageA,imageB):
    err=np.sum((imageA.astype("float")-imageB.astype("float"))**2)
    err /=float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
def compare_image(imageA,imageB,title):
    m=mse(imageA,imageB)
    s=ssim(imageA,imageB)
    fig=plt.figure(title)
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m,s))
    
    #show first image
    ax=fig.add_subplot(1,2,1)
    plt.imshow(imageA,cmap=plt.cm.gray)
    plt.axis("off")
    
    #show the second image
    ax=fig.add_subplot(1,2,2)
    plt.imshow(imageB,cmap=plt.cm.gray)
    plt.axis("off")
    
    plt.show()
```

```python
#loading the images
imag1=cv2.imread('camera-icon.png')
imag2=cv2.imread('camera-icon-output.png')
imag3=cv2.imread('canny edge output.png')
```

```python
#resizing the images look itno zero patting
imag1=cv2.resize(imag1,(10000,10000))
imag2=cv2.resize(imag2,(10000,10000))
imag3=cv2.resize(imag3,(10000,10000))
```

```python
#converting to grayscale cv2.cvtcolor(original,cv2.color_bgr2gray)
imag1=cv2.cvtColor(imag1,cv2.COLOR_BGR2GRAY)
imag2=cv2.cvtColor(imag2,cv2.COLOR_BGR2GRAY)
imag3=cv2.cvtColor(imag3,cv2.COLOR_BGR2GRAY)
```

```python
# Initialize the figure
fig = plt.figure("ssim comparison")
images = [("first image", imag1), ("second image", imag2), ("third image",imag3)]

# Loop over the images
for (i, (name, image)) in enumerate(images):
    # Show the image
    ax = fig.add_subplot(1, 3, i + 1)
    plt.imshow(image, cmap=plt.cm.gray)
    plt.axis("off")
    ax.set_title(name)  # Set the title to the image name

plt.show()
```

```python
#now to comapre betwwen images
compare_image(imag1,imag1,'same photo')
```

```python
compare_image(imag2,imag2,'same edge photo')
```

```python
compare_image(imag3,imag3,'same canny photo')
```

```python
compare_image(imag1,imag3, 'original image vs canny edge detected image')
```

```python
compare_image(imag1,imag2, 'original image vs quantum edge detected image')
```

```python
compare_image(imag3,imag2, 'canny image vs quantum edge detected image')
```
